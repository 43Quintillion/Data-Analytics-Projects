```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv(r'C:\Users\lambu\Documents\Data Analytics Stuff\Python\Movie Data Analysis\movies.csv')

df.head()
```

```python
# Checking for missing data 
print("A column's percentage of missing values:")
for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))


#The rows with missing values can't be filled in easily using given information and are not a big portion of the data
#Therefore, for this analysis rows with missing values will be dropped


# Copying into second data frame + dropping duplicates and missing values
df1 = df.drop_duplicates()
df1 = df.dropna() 
```

<h4>Data Cleaning</h4>

```python
df1.dtypes
```

```python
#Changing data types of some columns to more appropriate data types 

df1['budget'] = df1['budget'].astype('int64')
df1['gross'] = df1['gross'].astype('int64')
df1['votes'] = df1['votes'].astype('int64')
```

```python
#Looking at the data there are instances where the 'year' column does not match the year in the 'released' column. 
#Creating correct Year column based on the 'released' column

df1['yearcorrect'] = df1['released'].str.split(',').str[1].str.split().str[0]
df1
```

```python
# Ordering by gross column

df1 = df1.sort_values(by=['gross'], inplace = False, ascending = False)
```

<h4>Finding Correlations in the Data - Other Variables Against Gross Revenue</h4>

```python
# Boxplot - Looking for Outliers

df1.boxplot(column=['gross'])
```

```python
# budget vs gross scatter plot

plt.scatter(x=df['budget'], y = df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Film Budget')

plt.show()
```

```python
# budget vs gross regression plot using seaborn

sns.regplot(x='budget', y='gross', data=df1, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# Correlation Table with Numerical Fields Using .corr()

numeric_cols = df1.select_dtypes(include=['number']).columns

df1[numeric_cols].corr()

#Note: Highj correlation between budget and gross
```

```python
correlation_matrix = df1[numeric_cols].corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Analysing Company vs Gross

    #Label encoding text columns as numeric values

df_encoded = df1.copy()

for col_name in df_encoded.columns:
    if(df_encoded[col_name].dtype == 'object'):
        df_encoded[col_name] = df_encoded[col_name].astype('category')
        df_encoded[col_name] = df_encoded[col_name].cat.codes

df_encoded
```

```python
df1
```

```python
# Correlation matrix for the whole data set using label encoded data

correlation_matrix_whole = df_encoded.corr()

sns.heatmap(correlation_matrix_whole, annot=True)
plt.title('Correlation Matrix entire Dataset')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
